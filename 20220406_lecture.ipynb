* Variability Measures
  * Variance $\sigma^2 = \frac{\sum_{i=1}^n (x_i - \mu)^2}{n-1}$
    * Measurement of spread
    * Higher variance populations are farther from the mean on average
    * $n$ in denominator for a population, $n-1$ for a sample; in practice $n \approx n-1$, so it doesn't really matter.
  * Standard deviation $\sigma$; square root of the variance
  * If you know $\sigma^2$, you know $\sigma$, and vice-versa.
  * Some operations call for $\sigma$, some for $\sigma^2$.

```python
import pandas as pd
```

```python

# import csv into DataFrame
url = 'https://fw.cs.wwu.edu/~hutchib2/doc/AnnualEmployeeSalary2.csv'
df = pd.read_csv(url,thousands=',')
```

```python
# get some useful stats about df
df.info()
```

```python
# check the columns
df.columns
```

```python
# unique agencies
df['AgyTitle'].unique()
```

```python
# create a new dataframe with only WWU employees
wwu_df = df[df['AgyTitle'] == "Western Washington University"]
wwu_df
```

```python
# list unique job titles at WWU
wwu_df['Job Title'].unique()
```

```python
# list all job titles that include the word "professor" at WWU
profs = wwu_df['Job Title'].str.contains("PROFESSOR")
wwu_df[profs]['Job Title'].unique()
```

```python
# all sorts of summary statistics
wwu_df['Sal2020'].mean()
sal = wwu_df['Sal2020'] # save me from having to type wwu_df['Sal2020']
sal[sal>0].mean()
sal[sal>0].median()
sal[sal>0].mode()
wwu_df[wwu_df['Sal2020']==500]
sal.max()
sal.argmax()
wwu_df.iloc[6920]
sal[sal>0].min()
sal[sal>0].var()
sal[sal>0].std()
wwu_df.sort_values('Sal2020',ascending=False).iloc[0:20]
```
